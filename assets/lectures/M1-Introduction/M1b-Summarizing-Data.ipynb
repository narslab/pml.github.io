```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
cadence = [.95, .85, .92, .95, .93, .86, 1.00, .92, .85, .81, .78 ,.93 ,.93, 1.05, .93, 1.06, .96, .81, .96]
```

```python
plt.hist(cadence)
plt.xlabel('Cadence (strides per second')
plt.ylabel('Counts')
```

```python
m = np.mean(cadence)
med = np.median(cadence)
mod = stats.mode(cadence)
firstquart = np.percentile(cadence,25) 
```

```python
print(m, med, mod, firstquart)
```

```python
np.std(cadence)
```

```python
np.var(cadence)
```

```python
plt.boxplot(cadence)
```
